## Calculating implicit frequency

```python
vocab_words_underscored = [w.replace(" ", "_") for w in vocab_words]
vocab_words_underscored
```

```python
from gensim.models import KeyedVectors

import gensim.downloader as api

model = api.load("word2vec-google-news-300")  # downloads automatically (~1.6 GB)
```

```python
similar_words_dict = dict()

for w in vocab_words_underscored:
    try:
        similar_words = model.most_similar(w, topn=5)
        similar_words_dict[w] = similar_words
    except KeyError:
        # Word not in vocabulary
        similar_words_dict[w] = []
        print(f"Word '{w}' not found in Word2Vec vocabulary.")

# Print results
for word, sims in similar_words_dict.items():
    print(f"\nVocabulary word: {word}")
    if sims:
        for sim_word, score in sims:
            print(f"  {sim_word} ({score:.2f})")
    else:
        print(" No similar words found.")
```

```python
# Calculate implicit frequency
implicit_freq = {}
for word, synonyms in synonyms_dict.items():
    count = 0
    for syn in synonyms:
        count += text_lower.count(syn.lower())
    implicit_freq[word.lower()] = count

# Display results
for word, count in implicit_freq.items():
    print(f"{word}: {count}")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Words
words = list(explicit_counts.keys())
x = np.arange(len(words))  # positions on x-axis

# Bar width
width = 0.35

# Create plot
fig, ax = plt.subplots(figsize=(12, 6))

# Plot bars
rects1 = ax.bar(x - width/2, [explicit_counts[w] for w in words], width, label='Explicit', color='skyblue')
rects2 = ax.bar(x + width/2, [implicit_freq[w] for w in words], width, label='Implicit', color='orange')

# Labels and title
ax.set_ylabel('Frequency')
ax.set_xlabel('Vocabulary Words')
ax.set_title('Explicit vs Implicit Vocabulary Frequencies')
ax.set_xticks(x)
ax.set_xticklabels(words, rotation=45, ha='right')
ax.legend()

# Add frequency labels on top of bars
def add_labels(rects):
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()/2., height + 0.1,
                f'{int(height)}', ha='center', va='bottom', fontsize=10)

add_labels(rects1)
add_labels(rects2)

plt.tight_layout()
plt.show()
```
